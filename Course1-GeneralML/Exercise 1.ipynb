## Predicting House Prices

```python
houses = turicreate.SFrame('home_data.sframe')
houses
```

```python
turicreate.show(houses[1:5000]['sqft_living'], houses[1:5000]['price'])
```

```python
train_data, test_data = houses.random_split(.8, seed=0)
```

# Regression Model

```python
sqft_model = turicreate.linear_regression.create(train_data, target='price', features=['sqft_living'])
```

```python
sqft_model
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(test_data['sqft_living'],test_data['price'],'.',
        test_data['sqft_living'],sqft_model.predict(test_data),'-')
```
